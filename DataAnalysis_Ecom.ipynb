```python
import pandas as pd

# Step 1: Load the dataset
df = pd.read_csv("C:\\Users\\Varshini\\Downloads\\E-Commerce Survey (Responses) - Form responses 1 (1).csv")

# Step 2: Get the shape of the DataFrame
total_rows, total_cols = df.shape

# Step 3: Print the total number of rows and columns
print(f"Total number of rows: {total_rows}")
print(f"Total number of columns: {total_cols}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Visualize unique labels for each column using pie charts
for column in df.columns:
    unique_labels = df[column].unique()
    label_counts = df[column].value_counts()
    
    # If there are too many unique labels, skip visualization
    if len(unique_labels) > 10:
        print(f"Skipping visualization for '{column}' as there are too many unique labels ({len(unique_labels)})")
        continue
    
    plt.figure(figsize=(3, 3))
    plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'{column}')
    plt.axis('equal')
    plt.show()
```

# What each gender prefers?

```python
import pandas as pd


# Step 2: Check the number of unique labels in column 3 (gender)
gender_labels_count = df.iloc[:, 2].unique()

print(f"Number of unique labels in column 3 (gender): {gender_labels_count}")
# Step 3: Determine the most preferred choice for each gender from column 4 onwards
columns_of_interest = df.columns[3:]  # Starting from the 4th column

# Group by gender and find the mode for each column
gender_grouped = df.groupby(df.iloc[:, 2])

for gender, group_df in gender_grouped:
    print(f"\nGender: {gender}")
    for column in columns_of_interest:
        max_label = group_df[column].mode().iloc[0]  # Mode returns a Series, so we extract the first value
        print(f"{column}: {max_label}")
```

# preference based on Age

```python
import pandas as pd

age_column = df.iloc[:, 1] 
columns_of_interest = df.columns[2:] 

# Group by gender and find the mode for each column
age_grouped = df.groupby(age_column)

for age, group_df in age_grouped:
    print(f"\nage: {age}")
    for column in columns_of_interest:
        max_label = group_df[column].mode().iloc[0]  
        print(f"{column}: {max_label}")
```

# gender and age in our dataset

```python
import pandas as pd

# Step 2: Extract the column prior to the gender column (assuming age information)
age_column = df.iloc[:, 1]  # Assuming age is at index 1, adjust if necessary

# Step 3: Determine the most frequently occurring value for each gender
gender_column = df.iloc[:, 2]  # Assuming gender is at index 2
gender_labels = gender_column.unique()

for gender in gender_labels:
    print(f"\nGender: {gender}")
    filtered_age = age_column[gender_column == gender]
    most_common_age = filtered_age.mode().iloc[0] if not filtered_age.empty else None
    print(f"Most common age for {gender}: {most_common_age}")
```
